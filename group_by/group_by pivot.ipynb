С помощью метода groupby() данные были сгруппированы по цели кредита и количеству статуса домовладения. Подсчёт был сделан с помощью count.

Группировка - “employment_type”  и кпо каждой локации “company_location”. Результат должен быть выведен в следующем формате (на скриншоте представлен фрагмент).

С помощью Groupby создаётся датафрейм, в котором с помощью метода size и reset_index переименовывается столбец подсчета в Count, с помощью sort.values значения сортируются, с помощью ascending = False сортируются по убыванию.

```python
group1 = df.groupby(['Intent','Home'])
answer_group = group1.size().reset_index(name='count')
answer_group = answer_group.sort_values(by = 'count', ascending = False)
answer_group
```

По данной группировке можно сделать вывод, что большинство заявителей арендует жилье, также, наибольшее число заявителей берут кредит на образование. Наименьшее количество заявителей указали тип домовладения "Другой", и цель кредита - "Обучение" и "Личные цели".

Необходимо отфильтровать возраст до 100, создать новый датасет из них, выбрать цель кредита - education, personal, medical. Создать сводную таблицу по этим данным с любыми столбцами, но values = rate, aggfunc = mean. Создаём копию изначального датафрейма с помощью метода .copy(), с помощью метода .loc сортируем возраст, с помощью метода .isin() фильтруем цели кредита, после этого создаём сводную таблицу по новым данным.

```python
df2 = df.copy()
df2 = df.loc[df['Age'] <= 100]
Intents = ["MEDICAL", "EDUCATION", "PERSONAL"]
df2 = df2[df2['Intent'].isin(Intents)].reset_index()
pivot_dop = df2.pivot_table(index = 'Age', values = 'Rate', columns = 'Intent', aggfunc = 'mean')
pivot_dop
```

По данной сводной таблице можно сделать вывод, что наибольшая процентная ставка по кредиту у заявителей с возрастом 25 и целью кредита "Medical", наименьшая же у заявителей с возрастом 21 и целью кредита "Medical", в среднем же, по цели кредита "Education" процентная ставка - 11%, по "Medical" - 13%, по "Personal" - 12%.
